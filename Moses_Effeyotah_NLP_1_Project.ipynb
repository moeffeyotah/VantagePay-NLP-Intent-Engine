<div style="background-color: #1A1A1A; padding: 25px; border-left: 10px solid #E02035; font-family: 'Segoe UI', sans-serif; color: white;">
    <h2 style="color: #E02035; margin-top: 0; text-transform: uppercase; letter-spacing: 1px; font-weight: 800;">
        3. Data Preparation: NLP Pipeline & Vectorization
    </h2>
    <p style="font-size: 1.1em; line-height: 1.5;">
        This stage transforms raw unstructured text into a high-dimensional feature matrix. By executing a three-step pipeline, we reduce linguistic entropy and isolate the semantic intent:
    </p>
    <ul style="list-style-type: square; color: #E02035;">
        <li><span style="color: white;"><b>Linguistic Normalization:</b> Utilizing spaCy for <b>Lemmatization</b>, we map varied word forms (e.g., "ordering", "ordered") back to their dictionary roots to handle real-world "noise."</span></li>
        <li><span style="color: white;"><b>TF-IDF Vectorization:</b> We extracted <b>10,000 Dimensions</b> using Unigrams and Bigrams, ensuring that the model captures specific phrases like "cancel order" rather than just individual words.</span></li>
        <li><span style="color: white;"><b>Strategic Stratification:</b> The dataset was split into <b>Train (18,820)</b>, <b>Validation (4,021)</b>, and <b>Test (4,031)</b> sets to ensure rigorous model validation.</span></li>
    </ul>
    <p style="border-top: 1px solid rgba(255,255,255,0.1); padding-top: 10px; font-style: italic; color: #CCCCCC;">
        <b>Architect's Decision:</b> We expanded the feature space to 10,000 to capture the nuances of the 27 intent classes. Preprocessing is now complete, providing a clean, numerical foundation for the modeling phase.
    </p>
</div>

```python
import spacy 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.preprocessing import LabelEncoder 

# 1. Load spaCy for Preprocessing. 
#!python -m spacy download en_core_web_sm (Run if not already installed) 
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])

def lemmatize_text(text): 
    """Reduces words to their dictionary root and remove noise.""" 
    doc = nlp(text.lower()) 
    return " ".join([token.lemma_ for token in doc if token.is_alpha and not token.is_stop]) 

print("--- Step 1: Executing Lemmatization (Cleaning Noise) ---")
df['clean_instruction'] = df['instruction'].apply(lemmatize_text) 

# 2. Target Encoding. 
# Convert the 27 text intents into integers for the Neural Network. 
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['intent']) 

# 3. Feature Engineering: TF-IDF. 
print("--- Step 2: Feature Engineering (TF-IDF Vectorization) ---")
tfidf = TfidfVectorizer(max_features = 10000, ngram_range= (1, 2))
X_tfidf = tfidf.fit_transform(df['clean_instruction']) 

# 4. Dataset Splitting. 
# An initial split 80/20 Train-Test split, then split the 80 Train set into Train and Validation of 85/15 split.

print("-- Step 3: Dataset Stratification (Train/Validation/Test) ---")
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_tfidf, y, test_size = 0.15, random_state = 42, stratify = y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size = 0.176, random_state = 42, stratify =y_train_full 
)

print(f"\nPreprocessing Complete.")
print(f"Final Feature Space: {X_train.shape[1]} Dimensions") 
print(f"Training Samples: {X_train.shape[0]} | Validation: {X_val.shape[0]} | Test: {X_test.shape[0]}")
```

    <h2 style="color: #E02035; margin-top: 0; text-transform: uppercase; letter-spacing: 1px; font-weight: 800;">
        4. Modeling: Linear Support Vector Classification
    </h2>
    <p style="font-size: 1.1em; line-height: 1.5;">
        For high-dimensional text classification (10,000 features), we have deployed a <b>Linear Support Vector Classifier (SVC)</b>. Unlike tree-based models, SVCs are mathematically optimized for sparse TF-IDF matrices.
    </p>
    <ul style="list-style-type: square; color: #E02035;">
        <li><span style="color: white;"><b>High-Dimensional Geometry:</b> SVC finds the optimal hyperplane that maximizes the "margin" between 27 semantic intents.</span></li>
        <li><span style="color: white;"><b>Speed & Scalability:</b> The linear kernel provides millisecond inference times, essential for the VantagePay real-time microservice.</span></li>
        <li><span style="color: white;"><b>Regularization:</b> We utilize a 'Squared Hinge' loss function to penalize misclassifications, ensuring the model generalizes well to the unseen Test set.</span></li>
    </ul>
    <p style="border-top: 1px solid rgba(255,255,255,0.1); padding-top: 10px; font-style: italic; color: #CCCCCC;">
        <b>Architect's Decision:</b> We selected Linear SVC over Random Forest to prevent "overfitting" on the sparse feature space and to maintain a lightweight deployment footprint.
    </p>
</div>

**Stage 4: Baseline Model (Linear SVC)**

```python
import time
from sklearn.svm import LinearSVC 
from sklearn.metrics import classification_report, accuracy_score, f1_score 

# Baseline: Default LinearSVC 
model_baseline = LinearSVC(random_state = 42) 

start_time = time.time()
model_baseline.fit(X_train, y_train) 
train_time = time.time() - start_time 

# Predictions. 
y_val_pred = model_baseline.predict(X_val) 

# Metrics for the Table. 
validation_accuracy = accuracy_score(y_val, y_val_pred) 
validation_f1 = f1_score(y_val, y_val_pred, average = 'macro')

print(f"Baseline Complete.")
print(f"Validation Accuracy: {validation_accuracy:.4f}") 
print(f"Macro F1-Score: {validation_f1:.4f}")
print(f"Training Time: {train_time:.2f} seconds") 
```

```python
from sklearn.metrics import classification_report 
import numpy as np 

# 1. Full Diagnostic Report. 
print("=====================================================")
print("                CLASSIFICATION REPORT                ")
print("=====================================================")

# Target names are pulled directly from our label encoder. 
report = classification_report(y_val, y_val_pred, target_names = label_encoder.classes_) 
print(report) 

# 2. Programmatic Error Analysis (Finding the worst confusion) 
# Create a copy of the confusion matrix and zero out the correct guesses. 
cm_errors = cm.copy()
np.fill_diagonal(cm_errors, 0) 

# Find the coordinates of the highest number of errors. 
max_error_idx = np.unravel_index(np.argmax(cm_errors), cm_errors.shape) 
true_intent_failed = label_encoder.classes_[max_error_idx[0]] 
pred_intent_wrong = label_encoder.classes_[max_error_idx[1]] 
max_error_count = cm_errors[max_error_idx] 

print("\n=====================================================================")
print("                         CRITICAL ERROR ANALYSIS                       ")
print("=======================================================================")
if max_error_count > 0:
    print(f"Most Frequent Misclassification ({max_error_count} occurrences):") 
    print(f"-> The model confused Actual [{true_intent_failed}] as Predicted [{pred_intent_wrong}]") 
    print("\nGiven the 98.48% Macro F1, this semantic overlap is within acceptable risk tolerances.") 
else: 
    print("Zero miscclassifications detected. Flawless Execution.")   
```

```python
import joblib

# ----------------------------------------------
# 1. MODEL SERIALIZATION 
#-----------------------------------------------

print("Exporting Production Assets...") 
joblib.dump(model_baseline, 'fanshawe_intent_svc_model.pkl') 
joblib.dump(tfidf, 'fanshawe_tfidf_vectorizer.pkl') 
joblib.dump(label_encoder, 'fanshawe_label_encoder.pkl') 
print("✅ Assets successfully saved to disk (.pkl format).") 

#------------------------------------------------
# 2. THE INFERENCE PIPELINE
#------------------------------------------------

def predict_customer_intent(raw_text): 
    """
    Takes a raw string from a customer, cleans it, vectorzes it, and returns the predicted business intent.
    """

    # Step 1: Clean the text using the lemmatize_text function from the preprocessing stage. 
    clean_text = lemmatize_text(raw_text) 

    # Step 2: Vectorize the cleaned text using the saved TF-IDF vectorizer.
    vectorized_text = tfidf.transform([clean_text]) 

    # Step 3: Predict the intent using the saved SVC model.
    prediction_encoded = model_baseline.predict(vectorized_text) 

    # Step 4: Decode to human readble intents. 
    final_intent = label_encoder.inverse_transform(prediction_encoded)[0]
    return final_intent 

# --- Testing completely new data ---
print("\n==============================================")
print("        LIVE INFERENCE ENGINE TEST              ")
print("================================================")

test_sentences = [
    "Hey, I need my money back for order #12345, it arrived broken.",
    "Can someone help me change my password? I forgot it.",
    "Where is my package? It was supposed to be here yesterday." 
]

for sentence in test_sentences:
    prediction = predict_customer_intent(sentence) 
    print(f"Customer says: '{sentence}'") 
    print(f"-> Automated Routing: [{prediction.upper()}]\n") 

    
```

```python
import numpy as np

def get_top_trigger_words(intent_name, top_n=5):
    """Extracts the exact words the model uses to identify an intent."""
    # Find the index of the intent
    class_index = np.where(label_encoder.classes_ == intent_name)[0][0]
    
    # Get the coefficients (weights) for this specific class
    class_weights = model_baseline.coef_[class_index]
    
    # Match the weights to the TF-IDF vocabulary
    feature_names = tfidf.get_feature_names_out()
    
    # Sort by highest weight
    top_indices = np.argsort(class_weights)[-top_n:][::-1]
    top_words = [feature_names[i] for i in top_indices]
    
    return top_words

print("==========================================================")
print("             EXPLAINABLE AI: FEATURE WEIGHTS              ")
print("==========================================================")
# Let's look at the triggers for 'cancel_order' and 'get_refund'
print(f"Top 5 trigger words for [CANCEL_ORDER]: {get_top_trigger_words('cancel_order')}")
print(f"Top 5 trigger words for [GET_REFUND]: {get_top_trigger_words('get_refund')}")
print(f"Top 5 trigger words for [TRACK_ORDER]: {get_top_trigger_words('track_order')}")

```
